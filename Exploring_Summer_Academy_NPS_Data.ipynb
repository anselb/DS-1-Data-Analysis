```python
promoters = y2017_clean_df[(y2017_clean_df['Rating (Num)'] >= 9)].count()['ID']
passives = y2017_clean_df[(y2017_clean_df['Rating (Num)'] >= 7) & (y2017_clean_df['Rating (Num)'] <= 8)].count()['ID']
detractors = y2017_clean_df[(y2017_clean_df['Rating (Num)'] >= 1) & (y2017_clean_df['Rating (Num)'] <= 6)].count()['ID']
NPS = (promoters - detractors) / (promoters + passives + detractors) * 100
NPS
```

```python
for week in range(1, 8):
    print(y2017_clean_df.groupby(['Week']).get_group(week).mean())
```
