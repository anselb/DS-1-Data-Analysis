```python
# Creating new dataframe for macronutrient averages across all groups
protein_avg = []
fats_avg = []
carbs_avg = []

for item in nutrition['group'].unique():
    protein_avg.append(nutrition[nutrition['group'] == item].mean()['protein (g)'])
    fats_avg.append(nutrition[nutrition['group'] == item].mean()['fat (g)'])
    carbs_avg.append(nutrition[nutrition['group'] == item].mean()['carbohydrate (g)'])

raw_data = {'group': nutrition['group'].unique(),
        'protein_avg': protein_avg,
        'fats_avg': fats_avg,
        'carbs_avg': carbs_avg}
group_avgs = pd.DataFrame(raw_data, columns = ['group', 'protein_avg', 'fats_avg', 'carbs_avg'])

# Setting the positions and width for the bars
pos = list(range(len(nutrition['group'].unique()))) 
width = 0.25 

# Plotting the bars
fig, ax = plt.subplots(figsize=(10,5))

# Create a bar with protein data,
# in position pos,
plt.bar(pos, 
        #using group_avgs['protein_avg'] data,
        group_avgs['protein_avg'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#86F954', 
        # with label the first value in group
        label=group_avgs['group'][0])

# Create a bar with fats data,
# in position pos + some width buffer,
plt.bar([p + width for p in pos], 
        #using group_avgs['fats_avg'] data,
        group_avgs['fats_avg'],
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#EE3224', 
        # with label the second value in group
        label=group_avgs['group'][1]) 

# Create a bar with carbohydrates data,
# in position pos + some width buffer,
plt.bar([p + width*2 for p in pos], 
        #using group_avgs['carbs_avg'] data,
        group_avgs['carbs_avg'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#166DE0', 
        # with label the third value in group
        label=group_avgs['group'][2]) 

# Set the y axis label
ax.set_ylabel('Grams')

# Set the chart's title
ax.set_title('Macronutrient Averages per Group')

# Set the position of the x ticks
ax.set_xticks([p + 1.5 * width for p in pos])

# Set the labels for the x ticks
ax.set_xticklabels(group_avgs['group'], rotation='vertical')

# Setting the x-axis and y-axis limits
plt.xlim(min(pos)-width, max(pos)+width*4)
plt.ylim([0, max(group_avgs['protein_avg'] + group_avgs['fats_avg'] + group_avgs['carbs_avg'])] )

# Adding the legend and showing the plot
plt.legend(['Avg. Protein', 'Avg. Fat', 'Avg. Carbs.'], loc='upper left')
plt.grid()
plt.show()
```
